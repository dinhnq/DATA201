You are going to use PCA on the vector of the data (so the 768 long vector, not the $28 \times 28$ image). You can use either my PCA or the one from sklearn. 

Task 2: Use one of the PCA implementations and make a scree plot of the total explained variance for the first 100 principal components. Use that to decide how many principal components to use. (3 marks)

```python
# Here is PCA which is inspired from my instructor's source code.
def pca(data):
    # Compute the covariance matrix
    C = np.cov(data.T)

    # Get the eigenvalues and eigenvectors
    evals,evecs = np.linalg.eig(C)

    # Now need to sort them into descending order
    indices = np.argsort(evals)
    indices = indices[::-1]
    evecs = evecs[:,indices]
    evals = evals[indices]

    # Transform the data
    return np.dot(evecs.T,data.T).T, evals, evecs
```

```python
# Each datapoint consists of a unique ID, a class (number 0-9), 
# and then 768 ( =28×28 ) greyscale pixel values.
data_comps = data[:,2:]

# Center data
mean = np.mean(data_comps, axis=0)
new_data = data_comps - mean

data_100_comps, evals, evecs = pca(new_data)
data_100_comps = data_100_comps[:,:100]

cumsum_val = np.cumsum(evals)[:20]
# <=> np.sum(evals[:20])
sum_val = cumsum_val[-1]
explained_variances = cumsum_val / sum_val

print(data_100_comps.shape)

pl.plot(explained_variances)
```

Task 3: Compute the first two principal components, and plot the points in 2D with a different colour for each class. I used `import matplotlib.colors as mcolors` together with the list of colours: `cols = list(mcolors.TABLEAU_COLORS)`. Comment on the results. (3 marks)

```python
import matplotlib.colors as mcolors

colors = list(mcolors.TABLEAU_COLORS)

classes_col = data[:,1]

data_comps = data[:,2:]
mean = np.mean(data_comps, axis=0)
new_data = data_comps - mean

x_values, evals, evecs = pca(new_data)

component1 = x_values[:,0]
component2 = x_values[:,1]
```

```python
colors_size = len(colors)
for i in range(colors_size):
    indices = np.where(classes_col == classes[i])
    
    x_vals = component1[indices]
    y_vals = component2[indices]
    color_val = colors[i]
    
    #pl.plot(x_vals, y_vals, '.', color=color_val)
    #cheat
    pl.plot(x_vals*(-1), y_vals, '.', color=color_val)
```

```python
from sklearn.decomposition import PCA
pca = PCA(2)  # project from 64 to 2 dimensions

test_data = data[:,2:]
projected = pca.fit_transform(test_data)

component1 = projected[:,0]
component2 = projected[:,1]

for i in range(colors_size):
    indices = np.where(classes_col == classes[i])
    
    x_vals = component1[indices]
    y_vals = component2[indices]
    color_val = colors[i]
    
    pl.plot(x_vals, y_vals, '.', color=color_val)
pl.xlabel('component 1')    
pl.ylabel('component 2')
```
