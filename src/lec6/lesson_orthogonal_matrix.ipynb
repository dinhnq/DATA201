The determinant of any orthogonal matrix: det(Q) = +1 or -1.

(Dưới dạng) As a linear transformation, an orthogonal matrix (bảo toàn tích) preserves the inner product of vectors, and therefore acts as an isometry of Euclidean space, such as a rotation, reflection or rotoreflection. In other words, it's a unitary transformation.

The orthogonal group, O(n): the set of n x n orthogonal matrices

The special orthogonal group SO(n): the subgroup consisting orthogonal matrices with determinant +1. A special orthogonal matrix: each of its elements. As a linear transformation, every special orthogonal matrix acts as a rotation.

In mathematics, the conjugate transpose (or Hermitian transpose) of an m-by-n matrix A with complex entries is the n-by-m matrix obtained from A by taking the transpose and then taking the complex conjugate of each entry (the complex conjugate of a+ib being a-ib, for real numbers a and b). It is often denoted as $A^H$ or A*

For real matrices, the conjugate transpose is just the transpose, $A^H = A^T$

The conjugate transpose of an m x n matrix A: $(A^H)_{ij} = \bar A_{ji}$ or $A^H = (\bar A)^T = \bar{A^T}$

where $\bar A_{ji}$: a scalar complex conjugate with complex conjugated entries, and $A^T$: the transpose

A square matrix A with entries $a_{ij}$ is called:
- Hermitian or self-adjoint if $A = A^H$; i.e., $a_{ij} = \bar a_{ij}$
- Skew Hermitian or anti-hermitian if $A = -A^H$; i.e.,  $a_{ij} = -\bar a_{ij}$
- Normal if $A^HA = AA^H$
- Unitary if $A^H = A^{-1}$, equivalently $AA^H = I$, equivalently $A^HA = I$
